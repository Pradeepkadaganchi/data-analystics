```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zomato=pd.read_csv("Zomato.csv")
```

```python
#covert the column datatype in rating
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
zomato.head()
```

```python
zomato['rate']=zomato['rate'].apply(handleRate)
print(zomato.head())
```

```python
zomato.info()
```

```python
#Type of resturent
```

```python
sns.countplot(x=zomato['listed_in(type)'])
plt.xlabel('type of resturant')
```

```python
zomato.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Grouping data by 'listed_in(type)' and summing 'votes'
grouped_data = zomato.groupby('listed_in(type)')['votes'].sum()

# Converting grouped data into a DataFrame
result = grouped_data.reset_index()

# Plotting
plt.plot(result['listed_in(type)'], result['votes'], color="green", marker="o")
plt.xlabel('Type of Restaurant', color="red", size=20)
plt.ylabel('Total Votes', color="blue", size=20)
plt.title('Votes by Type of Restaurant', size=25)
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x="online_order",y="rate",data=zomato)
```

```python
pivot_table = zomato.pivot_table(index="listed_in(type)", columns="online_order", aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt="d")  # Correct colormap name
plt.title("Restaurant Types vs Online Order Availability")
plt.xlabel("Online Order")
plt.ylabel("Listed In(type)")
plt.show()
```
